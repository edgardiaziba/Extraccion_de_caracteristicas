#Limpieza de valores nulos y outliers.


```python
#carga desde un archivo .xlsx sin indices
Micro_Retailer=pd.read_excel('microretailer_mit_lift_lab.xlsx')
Micro_Retailer
```

```python
Micro_Retailer.info()
```

```python
#ajustar maximo de columnas
pd.options.display.max_columns= None
```

```python
columnas_numericas = Micro_Retailer.select_dtypes(include=['number'])
columnas_numericas.info()
```

```python
# Seleccionar columnas de tipo 'object' (categóricas)
columnas_categoricas = Micro_Retailer.select_dtypes(include=['object'])
columnas_categoricas
```

```python
columnas_categoricas.info()
```

```python
#Se imprimen los primeros 5 registros
columnas_categoricas.head(5)
```

```python
#Filtro de las 20 columnas por categoricas cualitativas y categoricas cuantitativas
CUANT=Micro_Retailer.iloc[ :,[14,16,21]]
CUAL=Micro_Retailer.iloc[ :,[3,10,11,12,15,20,30,35,37,38,39,40,46,47,48,50,53]]
```

```python
#Corroboramos valores nulos
valores_nulos=CUANT.isnull().sum()
valores_nulos
```

```python
CUANT=CUANT.fillna(method='ffill')
CUANT
```

```python
CUANT=CUANT.fillna(method='bfill')
```

```python
#Corroboramos valores nulos
valores_nulos=CUANT.isnull().sum()
valores_nulos
```

```python
y=CUANT
Limite_Superior = y.mean() + 3*y.std()
Limite_Inferior = y.mean() - 3*y.std()
print("Limite superior permitido: \n", Limite_Superior)
print("Limite inferior permitido: \n", Limite_Inferior)
```

```python
#Encontramos Outliers del data frame
outliers = CUANT[(y>Limite_Superior) | (y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
CUANT = CUANT[(y<=Limite_Superior) & (y>=Limite_Inferior)]
CUANT
```

```python
#Corroboramos valores nulos
valores_nulos=CUANT.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del data frame con "mean"
data_clean_CUANT=CUANT.copy()
data_clean_CUANT=data_clean_CUANT.fillna(round(CUANT.mean(), 1))
data_clean_CUANT
```

```python
#Corroboramos valores nulos
valores_nulos=data_clean_CUANT.isnull().sum()
valores_nulos
```

```python
#Corroboramos valores nulos
valores_nulos=CUAL.isnull().sum()
valores_nulos
```

```python
data_clean_CUAL=CUAL.copy()
data_clean_CUAL=data_clean_CUAL.fillna('DESCONOCIDO')
data_clean_CUAL
```

```python
#Corroboramos valores nulos
valores_nulos=data_clean_CUAL.isnull().sum()
valores_nulos
```

```python
#Unimos los data frames para crear un Micro_Retailer ya limpio y sin outliers
Micro_Retailer = pd.concat([data_clean_CUAL, data_clean_CUANT], axis = 1)
Micro_Retailer
```

```python
valores_nulos=Micro_Retailer.isnull().sum()
valores_nulos
```
